```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 70)

df = pd.read_csv("Issued_Construction_Permits.csv", low_memory=False)
df = df[df["Calendar Year Issued"].isin([2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013])]
```

```python
permits = df[["Permit Num", "Project ID", "TCAD ID", "Calendar Year Issued", "Status Current"]]
permits.sample(5)
```

```python
details = df[["Permit Num", "Issued Date", "Expires Date", "Completed Date", "Issuance Method", "Jurisdiction", "Council District"]]
details["Council District"] = details["Council District"].astype('Int64')
details.sample(5)
```

```python
projects = df[["Project ID", "Contractor Company Name", "Permit Type Desc", "Permit Class Mapped", "Work Class", "Total Job Valuation"]]
projects["Total Job Valuation"] = projects["Total Job Valuation"].astype('Int64')
projects.sample(5)
```

```python
contractors = df[["Contractor Company Name", "Contractor Full Name", "Contractor Trade", "Contractor Zip"]]
contractors = contractors.drop_duplicates()
contractors.sample(5)
```

```python
locations = df[["TCAD ID", "Applicant Full Name", "Original Zip", "Building Valuation", "Total Lot SQFT"]]
locations["Building Valuation"] = locations["Building Valuation"].astype('Int64')
locations["Total Lot SQFT"] = locations["Total Lot SQFT"].astype('Int64')
locations = locations.drop_duplicates()
locations.sample(5)
```

```python
applicants = df[["Applicant Full Name", "Applicant Organization", "Applicant Zip"]]
applicants = applicants.drop_duplicates()
applicants.sample(5)
```

```python
permits.to_csv("../elk/logstash/data/permits.csv", index=False)
details.to_csv("../elk/logstash/data/details.csv", index=False)

projects.to_csv("../elk/logstash/data/projects.csv", index=False)
contractors.to_csv("../elk/logstash/data/contractors.csv", index=False)

locations.to_csv("../elk/logstash/data/locations.csv", index=False)
applicants.to_csv("../elk/logstash/data/applicants.csv", index=False)
```
